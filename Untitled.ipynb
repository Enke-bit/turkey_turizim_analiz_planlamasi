# Bir ülke üzerinden tahmin işlemi yaptırma

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')
```

```python
# Tarih sütununu datetime formatına çevir ve zaman indeksi olarak ayarla
df['Date'] = pd.to_datetime(df['Date'])  # Doğru sütun adını kullan
df.set_index('Date', inplace=True)

# Almanya'dan gelen turist sayısını seç
germany_tourists = df[' ALMANYA']

# Veriyi görselleştir
plt.figure(figsize=(12, 6))
plt.plot(germany_tourists, label='Almanya')
plt.title('Türkiye\'ye Almanya\'dan Gelen Turist Sayısı')
plt.ylabel('Turist Sayısı')
plt.xlabel('Tarih')
plt.legend()
plt.show()

# Modeli oluştur ve eğit
model = ARIMA(germany_tourists, order=(5, 1, 0))
model_fit = model.fit(method_kwargs={"disp": 0})

# Model özetini yazdır
print(model_fit.summary())

# Gelecek 12 ay için tahmin yap
forecast = model_fit.forecast(steps=12)
print(forecast)

# Tahminleri görselleştir
plt.figure(figsize=(12, 6))
plt.plot(germany_tourists, label='Gerçek')
plt.plot(pd.date_range(start=germany_tourists.index[-1], periods=12, freq='M'), forecast, label='Tahmin')
plt.title('Gelecek 12 Ay İçin Almanya\'dan Gelen Turist Sayısı Tahmini')
plt.ylabel('Turist Sayısı')
plt.xlabel('Tarih')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
from dash import Dash, dcc, html, Input, Output
from statsmodels.tsa.arima.model import ARIMA
import pycountry
```

```python
# Sütun adlarını temizle (boşlukları kaldır, büyük harfe çevir)
df.columns = df.columns.str.strip().str.upper()

# Tarih sütununu datetime formatına çevir ve zaman indeksi olarak ayarla
df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)

# Dash uygulamasını başlat
app = Dash(__name__)

# Ülke isimlerini al
available_countries = df.columns.tolist()

# Ülke isimlerini ISO alpha-3 kodlarına çevir
def get_iso_alpha_3(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except LookupError:
        return None

map_data = pd.DataFrame({
    'country': available_countries,
    'iso_alpha': [get_iso_alpha_3(country) for country in available_countries]
})
map_data.dropna(inplace=True)

# Harita ve grafikleri içeren uygulama düzeni
app.layout = html.Div([
    dcc.Graph(id='world-map', config={'scrollZoom': False}),
    dcc.Graph(id='timeseries-plot'),
    dcc.Graph(id='forecast-plot'),
    dcc.Graph(id='bar-chart'),
])

# Haritayı güncelle
@app.callback(
    Output('world-map', 'figure'),
    Input('world-map', 'clickData')
)
def update_map(clickData):
    fig = px.scatter_geo(map_data, locations='iso_alpha', hover_name='country')
    fig.update_layout(title='Ülke Seçimi için Tıklayın', geo=dict(showframe=False, showcoastlines=False))
    return fig

# Grafikleri güncelle
@app.callback(
    [Output('timeseries-plot', 'figure'),
     Output('forecast-plot', 'figure'),
     Output('bar-chart', 'figure')],
    [Input('world-map', 'clickData')]
)
def update_graphs(clickData):
    if clickData is None:
        selected_country = 'ALMANYA'  # Varsayılan ülke
    else:
        selected_country = clickData['points'][0]['hovertext']

    # Seçilen ülkenin verilerini al
    country_tourists = df[selected_country]

    # Zaman serisi grafiği
    fig1 = go.Figure()
    fig1.add_trace(go.Scatter(x=country_tourists.index, y=country_tourists, mode='lines', name='Gerçek'))
    fig1.update_layout(title=f"{selected_country} Ülkesinden Gelen Turist Sayısı", xaxis_title='Tarih', yaxis_title='Turist Sayısı')

    # Modeli oluştur ve eğit
    model = ARIMA(country_tourists, order=(5, 1, 0))
    model_fit = model.fit()

    # Gelecek 12 ay için tahmin yap
    forecast = model_fit.forecast(steps=12)

    # Tahmin grafiği
    fig2 = go.Figure()
    fig2.add_trace(go.Scatter(x=country_tourists.index, y=country_tourists, mode='lines', name='Gerçek'))
    fig2.add_trace(go.Scatter(x=pd.date_range(start=country_tourists.index[-1], periods=12, freq='M'), y=forecast, mode='lines', name='Tahmin'))
    fig2.update_layout(title=f"Gelecek 12 Ay İçin {selected_country} Ülkesinden Gelen Turist Sayısı Tahmini", xaxis_title='Tarih', yaxis_title='Turist Sayısı')

    # Bar grafiği
    bar_data = df.sum().sort_values(ascending=False)
    fig3 = go.Figure([go.Bar(x=bar_data.index, y=bar_data.values)])
    fig3.update_layout(title='Ülkelere Göre Toplam Turist Sayısı', xaxis_title='Ülke', yaxis_title='Toplam Turist Sayısı')

    return fig1, fig2, fig3

if __name__ == '__main__':
    app.run_server(debug=True)
```
